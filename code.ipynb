```python
import tkinter as tk
import random

# Tkinter setup
root = tk.Tk()
root.title("Blackjack Display")
root.configure(bg = 'green')
root.geometry('800x150')

# frame to hold the player's cards
player_frame = tk.Frame(root, bg = 'gray')
player_frame.pack(side='bottom', pady=10)

# frame to hold the dealer's cards
dealer_frame = tk.Frame(root, bg='gray')
dealer_frame.pack(side='top', pady=10)

def add_card_to_frame(card, frame):
    card_label = tk.Label(frame, text=card)
    card_label.pack(side='left', padx=5)
    root.update()
    
def clear_frames():
    for widget in player_frame.winfo_children():
        widget.destroy()
    for widget in dealer_frame.winfo_children():
        widget.destroy()
        
# create a deck
deck = []
def reset():
    suits = ['Spades', 'Hearts', 'Clubs', 'Diamonds']
    royals = ['Jack','Queen','King','Ace']
    for suit in suits:
        # numbers
        for num in range(9):
            deck.append(f"{num+2} of {suit}")
        # royals
        for royal in royals:
            deck.append(f"{royal} of {suit}")
    # if casino uses 8 decks
    return deck*8

def section():
    print('_' * 10)

# calculate function
def calculate():
    global deck
    global player_tot
    global player_hand
    calc = input("Would you like to \n1.) Calculate the probability of a speicific value (Y/N) \n2.) Calculate the chance you won\'t' bust \n3.) Don\'t calculate \n Enter 1, 2, or 3: ")
    if calc == '1':
        card = input('Which number would you like to find the probability of getting? ').upper()
        ss = float(len(deck))
        # if a royal
        if card[0] in ['1','K','Q','J']:
            es = float(sum([1 for x in deck if x[0] in ['J','Q','K','1']]))
        else:
            es = float(sum([card[0:1] in x for x in deck]))
        print(f"Sample Space: {ss}, number of outcomes (cards left in the deck).")
        print(f"Event Space: {es}, number of successfull outcomes ({card}s left).")
        print("To calculate, we divide the event space by the sample space to get:")
        print(f"{es/ss:.3f} = to a {es/ss*100:.3f}% chance.")
    if calc == '2':
        # calculating temporary total for if Ace is in hand
        temp_tot = 0
        for card in player_hand:
            if card[0] == 'A':
                temp_tot += 1
            elif card[0] in ['J','Q','K','1']:
                temp_tot += 10
            else:
                temp_tot += int(card[0])
                
        max_value = 21 - temp_tot
        ss = float(len(deck))
        if max_value >= 10:
            es = ss
            max_value = 10
        else:
            # number of aces
            aces = float(sum([1 for card in deck if card[0] == 'A']))
            # number cards less than max_value but greater than 1
            num = 0
            i = 2
            while i <= max_value:
                for card in deck:
                    if card[0] == str(i):
                        num += 1
                i += 1
            es = num + aces
        
        print(f"Sample Space: {ss}, number of outcomes (cards left in the deck).")
        print(f"Event Space: {es}, number of successfull outcomes (cards less than or equal to {max_value}).")
        print("To calculate, we divide the event space by the sample space to get:")
        print(f"{es/ss:.3f} = to a {es/ss*100:.3f}% chance.")



def deal(string):
    # deal card
    global player_hand
    global dealer_hand
    card = random.choice(deck)
    deck.remove(card)
    if string == 'you':
        add_card_to_frame(card, player_frame) 
    else:
        add_card_to_frame(card,dealer_frame)
    print(f"\n{string.capitalize()} got the card {card}.")
    # add card to hand
    if string == 'you':
        player_hand.append(card)
        total(player_hand,'you')
        print("You now have", ', '.join(map(str,player_hand)),f':\na total of {player_tot}.\n')
    else:
        dealer_hand.append(card)
        total(dealer_hand,'dealer')
        print(f"Dealer now has", ', '.join(map(str,dealer_hand)),f':\na total of {dealer_tot}.\n')

def total(hand, turn):
    global player_tot
    global dealer_tot
    global player_hand
    global dealer_hand
    # convert to number
    
    if turn == 'you':
        player_tot = 0
        ace_count= 0
        for card in hand:
            if card[0] == 'A':
                ace_count += 1
                player_tot += 11
            elif card[0] in ['J','Q','K','1']:
                player_tot += 10
            else:
                player_tot += int(card[0])
        # accounting for multiple aces        
        while ace_count > 0 and player_tot > 21:
            player_tot -= 10
            ace_count -= 1     
    else:
        dealer_tot = 0
        ace_count= 0
        for card in hand:
            if card[0] == 'A':
                ace_count += 1
                dealer_tot += 11
            elif card[0] in ['J','Q','K','1']:
                dealer_tot += 10
            else:
                dealer_tot += int(card[0])
            
        # accounting for multiple aces        
        while ace_count > 0 and dealer_tot > 21:
            dealer_tot -= 10
            ace_count -= 1 
        



print('Welcome to Black Jack!')
money = round(float(input('How many pounds would you like to buy in for? ')),1)

will = 'Y'
deck = reset()
while will == 'Y':
    clear_frames()
    player_hand = []
    dealer_hand = []
    player_tot = 0
    dealer_tot = 0
    
    if money == 0:
        money += float(input('Uh oh! It looks like you are out of money. How much more would you like to buy in for? '))
    
    bet = float(input('How much money would you like to put in this round? '))
    if bet > money:
        bet = float(input(f"You don\'t have enough. You only have {money} pounds. Enter a different amount: "))
        

    # first card to you
    deal('you')
    
    
    # second card to dealer
    deal('dealer')
    

    # third card to you
    deal('you')
  

    # option to double
    total(player_hand, 'you')
    if player_tot == 21:
        money += (bet*3/2)
        print('Congratulations! You got Blackjack.')
        print(f'You won 3/2 times the normal win {bet*3/2}')
    else:
        double = 'N'
        if money/2 >= bet:
            double = input('Would you like to Double Down (Y/N)? ').upper()

        # user hit or stay
        if double == 'Y':
            section()
            bet = bet * 2
            # fifth card to you
            deal('you')
        else:
            calculate()
            hit = input('Would you like to hit or stay? ').lower()
            while hit == 'hit':
                # fifth card to you
                deal('you')
                total(player_hand, 'you')
                if player_tot < 21:
                    calculate()
                    hit = input('Would you like to hit or stay? ').lower()
                else:
                    hit = 'stay'
        section()

        if player_tot > 21:
            money -= bet
            print('Sorry you bust!')
            print(f'You lost {bet} pounds. Unlucky to be honest.')
        else:
            # fourth card to dealer
            deal('dealer')

            total(dealer_hand, 'dealer')
            # dealer hit or stay
            while dealer_tot < 17:
                print('Because dealer is under 17, he has to hit.')
                deal('dealer')
                total(dealer_hand, 'dealer')

            # settle scores
            section()
            if dealer_tot > 21:
                money += bet
                print('The Dealer bust!!')
                print(f'Congrats! You won {bet} pounds this round.')
            else:    
                if player_tot > dealer_tot:
                    money += bet
                    print(f"You have {player_tot} and the dealer has {dealer_tot}")
                    print(f'Congrats! You won {bet} pounds this round.')
                elif player_tot < dealer_tot:
                    money -= bet
                    print(f"You have {player_tot} and the dealer has {dealer_tot}")
                    print(f'Sorry! You lost {bet} pounds this round. Unlucky to be honest. You got it next round!')
                else:
                    print("You pushed with the dealer, so you neither won nor lost.")   


    
    print(f'You now have {money} pounds.')
    will = input('Would you like to continue playing (Y/N)? ').upper()
    
root.mainloop()

print('Thanks for playing with us today! And remember the house never wins.')




    
```
